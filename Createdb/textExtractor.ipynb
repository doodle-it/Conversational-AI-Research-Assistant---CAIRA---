```python
def remove_latex_symbols(s : str) -> str:
    # doc = fitz.open(pdf_path)
    text = ""
    # for page in doc:
    # Remove LaTeX content using regular expressions
    cleaned_text = re.sub(r'\$.*?\$', '', s)
    text += cleaned_text + "\n"
    return text
```

```python
from langchain_community.document_loaders import UnstructuredFileLoader
from unstructured.cleaners.core import clean_extra_whitespace

papers = []

for file in files:
    
    loader = UnstructuredFileLoader(file, post_processors=[clean_extra_whitespace])
    doc = loader.load()
    papers.append(doc)
```
